```python
import pandas as pd
```

```python
# Reading the dataset
df = pd.read_csv("employee.csv")

# checking the head
df.head()
```

```python
# Selecting columns with integer data type from the DataFrame
df.select_dtypes("int")
```

```python
df.select_dtypes("object")
```

```python
# Finding the maximum value in the MonthlyIncome column
print(df["MonthlyIncome"].max())

# Finding the minimum value in the MonthlyIncome column
print(df["MonthlyIncome"].min())
```

```python
# Calculate measures of central tendency
mean_salary = df['MonthlyIncome'].mean()
median_salary = df['MonthlyIncome'].median()
mode_salary = df['MonthlyIncome'].mode().values

print("Measures of Central Tendency - MonthlyIncome:")
print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
```

```python
# Calculate measures of spread
salary_range = df['MonthlyIncome'].max() - df['MonthlyIncome'].min()
salary_variance = df['MonthlyIncome'].var()
salary_std_dev = df['MonthlyIncome'].std()

print("Measures of Spread - MonthlyIncome:")
print("Salary Range:", salary_range)
print("Salary Variance:", salary_variance)
print("Salary Standard Deviation:", salary_std_dev)
```

## Measures of Dependence

To calculate measures of dependence or correlation between variables in employee data using Python, you can use the pandas library. The most common measure of dependence is the correlation coefficient, which quantifies the strength and direction of the linear relationship between two numerical variables. Here's an example of how to calculate the correlation coefficient on employee salary and years of experience data:

```python
# Calculate correlation coefficient
correlation_coefficient = df['MonthlyIncome'].corr(df['TotalWorkingYears'])

print("Correlation Coefficient:")
print(correlation_coefficient)
```

The resulting correlation coefficient ranges from -1 to 1, with values closer to -1 indicating a strong negative correlation, values closer to 1 indicating a strong positive correlation, and values close to 0 indicating little to no linear relationship.

By applying the corr() method, you can calculate the correlation coefficient to assess the dependence or relationship between different numerical variables in the employee dataset.
